```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import r2_score
```

## Datos:

```python
df = pd.read_csv('housing.csv')
df['total_bedrooms'] = df['total_bedrooms'].fillna(0)
df.head()
```

```python
plt.scatter(df.median_income, df.median_house_value)
```

```python
df = df[df.median_house_value != df.median_house_value.max()]
plt.scatter(df.median_house_value , df.median_income)
```

## Regresiones lineales:

```python
# housing median age
X = df[['housing_median_age']]
y = df['median_house_value']

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

```python
# total rooms
X = df[['total_rooms']]

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

```python
# total bedrooms
X = df[['total_bedrooms']]

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

```python
# population
X = df[['population']]

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

```python
# households
X = df[['households']]

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

```python
# median income
X = df[['median_income']]

model = linear_model.LinearRegression().fit(X, y)

print(r2_score(y_true = y, y_pred = model.predict(X)))
```

## An√°lisis:

- Con las regresiones lineales hechas anteriormente, podemos ver que el modelo hecho con la variable num√©rica de **median income** es el que tiene mayor precisi√≥n, lo cu√°l nos permite tener una mejor predicci√≥n en cuesti√≥n del valor futuro de las casas. Es decir, esta variable cuenta con un valor de **R2 de 0.4129**, el cu√°l nos indica que el modelo hecho con median income tiene una precisi√≥n de aproximadamente 41%, siendo as√≠ el mejor modelo para realizar.
